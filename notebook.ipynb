```python
import pandas as pd
```

We start with exploring the content of the dataset.

```python
df = pd.read_csv("Scotland_teaching_file_1PCT.csv")
```

```python
df
```

```python
# Remove first two columns, which contain useless data
df = df.drop(["Record_Number", "Region"], axis=1)

df
```

```python
# Variable stores a list of column names
cols = df.columns.tolist()

# Print distribution of values in each column
for i in range(len(cols)):
    print(df[cols[i]].value_counts().sort_index())
```

```python
# Count how many duplicate rows there are,
# the "size" column shows number of identical columns
df.groupby(cols, as_index=False).size()
```
